```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("DEvideos.csv", encoding='ISO-8859-1')
df = df.drop_duplicates(subset='video_id')
df = df.dropna(subset=['title', 'tags', 'views', 'likes', 'comment_count'])

# Parse dates and engineer time-based features
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
df['publish_day'] = df['publish_time'].dt.dayofweek
df['publish_hour'] = df['publish_time'].dt.hour

# Target variable: Viral if views > 500K
df['viral'] = (df['views'] > 500_000).astype(int)

# Encode category_id
le = LabelEncoder()
df['category_id'] = le.fit_transform(df['category_id'].astype(str))

# TF-IDF on video title
tfidf = TfidfVectorizer(max_features=100)
title_tfidf = tfidf.fit_transform(df['title']).toarray()
title_tfidf_df = pd.DataFrame(title_tfidf, columns=[f"title_tfidf_{i}" for i in range(title_tfidf.shape[1])])

# Combine all features
final_df = pd.concat([
    title_tfidf_df,
    df[['likes', 'comment_count', 'category_id', 'publish_day', 'publish_hour']].reset_index(drop=True)
], axis=1)

# Convert all column names to string
final_df.columns = final_df.columns.astype(str)

# Target
y = df['viral']

# Split data
X_train, X_test, y_train, y_test = train_test_split(final_df, y, test_size=0.2, random_state=42)

# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict
y_pred = rf_model.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Viral", "Viral"])
disp.plot(cmap="Greens")
plt.title("Random Forest - Confusion Matrix")
plt.show()

# Classification Report
print("Random Forest Model Report:")
print(classification_report(y_test, y_pred, target_names=["Not Viral", "Viral"]))

# Optional: Show top feature importances
importances = rf_model.feature_importances_
feature_names = final_df.columns
importance_df = pd.DataFrame({"Feature": feature_names, "Importance": importances})
importance_df = importance_df.sort_values(by="Importance", ascending=False).head(10)

print("\nTop 10 Important Features:")
print(importance_df)
```
